### 태그와 짧은 내용 같이 담기

```python
def add_tag_and_short_info_with(add_df):
    # add_df 정보 먼저 담기
    add_df_data = {}
    for idx in range(len(add_df)):
        if type(add_df.iloc[idx]['info']) ==float:
            add_df_data[add_df.iloc[idx]['place']] = add_df.iloc[idx]['태그'].strip()
        elif type(add_df.iloc[idx]['태그'])==float:
            add_df_data[add_df.iloc[idx]['place']] = add_df.iloc[idx]['info'].strip()
        else:
            add_df_data[add_df.iloc[idx]['place']] = add_df.iloc[idx]['info'].strip() + " " + add_df.iloc[idx]['태그'].strip()
    
    # {장소이름 : "인포 태그"} 로 저장
    return add_df_data
```

```python
add_df_data = add_tag_and_short_info_with(add_df)
add_df_data
```

```python
def add_to_csv_with(add_df_data,tour_df):
    additional_list = []

    for place in list(tour_df['place']):
        if place in add_df_data:
            additional_list.append(add_df_data[place])
        else:
            additional_list.append(None)    # None으로 넣어주기
    tour_df['태그'] = additional_list       # 태그라는 컬럼에 같이 넣어주기
    return tour_df
```

### 태그와 짧은 내용 따로 담기


```python
def add_tag_and_short_info_sep(add_df):
    # add_df 정보 먼저 담기
    add_df_data = {}
    for idx in range(len(add_df)):
        if type(add_df.iloc[idx]['info']) ==float:
            add_df_data[add_df.iloc[idx]['place']] = {'태그': add_df.iloc[idx]['태그'].strip(), 'short_info':''}
        elif type(add_df.iloc[idx]['태그']) ==float:
            add_df_data[add_df.iloc[idx]['place']] = {'태그': '', 'short_info': add_df.iloc[idx]['info']}
        else:
            add_df_data[add_df.iloc[idx]['place']] = {'태그': add_df.iloc[idx]['태그'].strip(), 'short_info': add_df.iloc[idx]['info'].strip()}
    return add_df_data
    # {장소이름 : { 태그 : 태그들, 'short_info' : 짧은 설명}} 로 저장
```

```python
add_df_data = add_tag_and_short_info_sep(add_df)
add_df_data
```

```python
def add_to_csv(add_df_data,tour_df):
    tag_list = []
    short_info_list = []

    for place in list(tour_df['place']):
        if place in add_df_data:
            tag_list.append(add_df_data[place]['태그'])
            short_info_list.append(add_df_data[place]['short_info'])
        else:
            tag_list.append(None)       # None으로 넣어주기
            short_info_list.append(None)    # None으로 넣어주기
    tour_df['태그'] = tag_list
    tour_df['short_info'] = short_info_list
    return tour_df
```

```python
tour_df = add_to_csv(add_df_data,tour_df)
tour_df.info()
```
